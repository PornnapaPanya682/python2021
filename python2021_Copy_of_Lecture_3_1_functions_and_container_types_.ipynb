##Task 2 The Half Adder

In  a digital circuit, the number is only “0” or “1”.
Wnen we add, if we add without carry on digit. It is the half adder, the half adder receive two digit value, and return the output (s as an sum bit), and c the carry on  (c as carry on bit) from the adder. 
the block diagram of the half adder as given

image.png

Making half adder circuit that can find the result of the summation of two digits using logic gates (AND, OR, NOT, NOR, NAND, XOR)




**Hint**

1. create the function which receive two digit value, and return the output of each of logic gate AND, OR, and NOT
2. NOR and NAND is the not Or, and Not And. So you can create the functions from AND, OR, NOT function
3. using the methods to crate a new half adder functions
The half adder is created using this circuit
image.png


4. as there are two outputs, two functions are required, one for sum bit output, and the other for carry on bit output



````


a = int(input("A: "))
b = int(input("B: "))

# nand function
def nand (a,b):
  if a==1 and b==1:
    return 0
  else:
    return 1

n1 = nand(a,b)
n2 = nand(a,n1)
n3 = nand(n1,b)
sum = nand(n2,n3)
carry = nand(n1,n1)

print("nand gate ==> sum: ", sum," carry : ", carry)

# truth table
def sum(a,b):
  if a==b:
    return 0
  else:
    return 1

def carry (a,b):
  if a==1 and b==1:
    return 1
  else :
    return 0

s = sum(a,b)
c = carry(a,b)

print("truth tabel ==> sum: ", s," carry : ", c)
````

##Task 3 Full Adder
from the half adder function, create the full adder. The full adder is to added to data, wit the carry in input. the block diagram is shown as given

image.png


````
a = int(input("A: "))
b = int(input("B: "))
c_in = int(input("Carry in: "))

# nand function
def nand (a,b):
  if a==1 and b==1:
    return 0
  else:
    return 1

def orGate(a,b):
  if a==0 and b==0:
    return 0
  else:
    return 1

# half
n1 = nand(a,b)
n2 = nand(a,n1)
n3 = nand(n1,b)
sum1 = nand(n2,n3)
c_out1 = nand(n1,n1)
# full
n1 = nand(sum1,c_in)
n2 = nand(sum1,n1)
n3 = nand(n1,c_in)
sum2 = nand(n2,n3)
c_out2 = nand(n1,n1)

carry_out=orGate(c_out1,c_out2)

print ("Way1 => s=",sum2," c=",carry_out)

# -----------------------------------
def half(a,b):
  n1 = nand(a,b)
  n2 = nand(a,n1)
  n3 = nand(n1,b)
  sum = nand(n2,n3)
  c_out = nand(n1,n1)
  return (sum,c_out)

h1=half(a,b)
h2=half(h1[0],h1[1])
carry_out=orGate(h1[1],h2[1])
print ("Way2 => s=", h2[0]," c=",carry_out)
````
